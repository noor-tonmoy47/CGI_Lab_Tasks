# Bresenham's Line Drawing Algorithm
### Below I write function for Bresenham
This algorithm will work for `0 < slope < 1` as it goes through x - axis and in slope < 1 the x - axis is dominent and we increase x axis by 1 in each increment.
When `slope>1` I swap the `(x1,y1) with (y1,x1)` and `(x2,y2) and (y2,x2)`.
So, bressman written below for this slope is given below:

```python
def bresenham(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    slope = dy > dx

    if slope:
        x1, y1 = y1, x1
        x2, y2 = y2, x2

    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    error = dx // 2
    y = y1
    yforward = 1 if y1 < y2 else -1

    points = []
    for x in range(x1, x2 + 1):
        coord = (y, x) if slope else (x, y)
        points.append(coord)
        error -= dy
        if error < 0:
            y += yforward
            error += dx

    return points
```

## case-1: m>1 `(1,1) and (4,8)`

```python
import matplotlib.pyplot as plt

# Define the endpoints of the line where m>1
x0, y0 = 1, 1
x1, y1 = 4, 8

grid = bresenham(x0, y0, x1, y1)

for point in grid:
    print(point)

plt.plot(*zip(*grid), color='red', marker='o', linestyle='None')

plt.title('Bresenham Line Drawing Algorithm')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, color='black', linewidth=1, linestyle='-')

plt.show()
```
